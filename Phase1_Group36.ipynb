## Introduction <a id='intro'></a>

### Dataset Source <a id='data_source'></a>

The Airline Passenger Satisfaction dataset used in our report has been extracted from Dr. Vural Aksakalli's [github page](https://github.com/vaksakalli/datasets/blob/master/airline_passenger_satisfaction.csv) which was further sourced from [Kaggle](https://www.kaggle.com/teejmahal20/airline-passenger-satisfaction). The dataset contains an airline passenger satisfaction survey. The goal is to predict customer satisfaction levels.

### Dataset Details <a id='data_details'></a>
This dataset is about a survey conducted to gain knowledge and insights into airline passengers' satisfaction. This contains information about type of customer, purpose of travel, gender, flight class, distance to be traveled, inflight services like food and drinks, seat comfort, entertainment system, leg room service, cleanliness and their overall satisfaction or dissatisfaction.

This dataset has a total of 23 variables and 129880 observations.

#### Finding outliers 
Next, we find out the outliers for all the integer variables in our dataset.

Checking for outliers in Age:

```python
q1_val = df['age'].quantile(0.25)
q2_val = df['age'].quantile(0.50) # this is also the median
q3_val = df['age'].quantile(0.75)
iqr = q3_val - q1_val
max_whisker = q3_val + 1.5*iqr
min_whisker = q1_val - 1.5*iqr
age_outliers = df[(df['age'] < min_whisker) | (df['age'] > max_whisker)]['age']
print('Number of "age" outliers:', age_outliers.shape[0])
df = df.drop(age_outliers.index)
plt.figure(figsize = (7,5))
sns.boxplot(df['age']);
plt.title('Boxplot of Passenger Age', fontsize = 15)
plt.xlabel('Age',color='black')
plt.show();
```

<br>Checking for outliers in flight distance:

```python
q1_val = df['flight_distance'].quantile(0.25)
q2_val = df['flight_distance'].quantile(0.50) # this is also the median
q3_val = df['flight_distance'].quantile(0.75)
iqr = q3_val - q1_val
max_whisker = q3_val + 1.5*iqr
min_whisker = q1_val - 1.5*iqr
flight_distance_outliers = df[(df['flight_distance'] < min_whisker) | (df['flight_distance'] > max_whisker)]['flight_distance']
print('Number of "flight_distance" outliers:', flight_distance_outliers.shape[0])
plt.figure(figsize = (7,5))
sns.boxplot(df['flight_distance']);
plt.title('Boxplot of Flight Distance', fontsize = 15)
plt.xlabel('Flight Distance', color='black')
plt.show();

df = df.drop(flight_distance_outliers.index)
print()
print('Box plot after dropping all the rows with outliers:')
plt.figure(figsize = (7,5))
sns.boxplot(df['flight_distance']);
plt.title('Boxplot of Flight Distance', fontsize = 15)
plt.xlabel('Flight Distance', color='black')
plt.show();
```

**Outliers based on scale rating of 1 to 5:**

```python
general_outliers = df[(df['inflight_wifi_service'] < 1) | (df['inflight_wifi_service'] > 5) |
                      (df['gate_location'] < 1) | (df['gate_location'] > 5) |
                      (df['food_and_drink'] < 1) | (df['food_and_drink'] > 5) | 
                      (df['departure_arrival_time_convenience'] < 1) | (df['departure_arrival_time_convenience'] > 5)|
                      (df['ease_of_online_booking'] < 1) | (df['ease_of_online_booking'] > 5)| 
                      (df['online_boarding'] < 1) | (df['online_boarding'] > 5)| 
                      (df['seat_comfort'] < 1) | (df['seat_comfort'] > 5)| 
                      (df['inflight_entertainment'] < 1) | (df['inflight_entertainment'] > 5)|  
                      (df['leg_room_service'] < 1) | (df['leg_room_service'] > 5)|
                      (df['baggage_handling'] < 1) | (df['baggage_handling'] > 5)| 
                      (df['checkin_service'] < 1) | (df['checkin_service'] > 5)| 
                      (df['inflight_service'] < 1) | (df['inflight_service'] > 5)| 
                      (df['cleanliness'] < 1) | (df['cleanliness'] > 5)]
print('Number of general outliers:', general_outliers.shape[0])    
df = df.drop(general_outliers.index)
print('The shape of the dataset after dropping all the outliers is: ',df.shape)
```

## Data Exploration & Visualisation <a id='data_explore_visual'></a>

### Univariate Visualisation <a id='uni_var'></a>

**1. Customer Class vs Number of People**
- In the following bar graph, we have depicted the count of people who travel on the 3 main seat classes in a flight namely Economy, Economy Plus and Business. 

- We have plotted this graph to get an estimate of which class people prefer to travel on the most. 

```python
plt.figure(figsize = (7,5))
fig = sns.countplot(x = 'customer_class', data = df, palette = 'rocket', 
                    order = df['customer_class'].value_counts().index)
fig = plt.title('Bar graph for Customer Class', fontsize = 15)
plt.xlabel('Customer Class', color='black')
plt.ylabel('Number of People', color='black')
plt.show()
```

- Here, we observe that the frequency of people travelling on Business Class slightly surpasses Economy Class. Whereas, Economy Plus Class was the least preferred.  

**2. Food and Drink vs Number of People**
- In the following Dot Plot, we have depicted the rating that people gave for food and drinks. 

- We have plotted this graph to get an idea of how satisfied people were with the food and drinks they got onboard.

```python
data=df['food_and_drink']
pos = [] 
keys = {} # this dict will help to keep track ...

# this loop will give us a list of frequencies to each number
for num in data: 
   if num not in keys:
      keys[num] = 1
      pos.append(1)
   else:
      keys[num] += 1
      pos.append(keys[num])

plt.scatter(data, pos)
plt.title('Dot Plot for Food and Drink Ratings vs the Number of People')
plt.xlabel('Food and Drink Rating (1-5)', color='black')
plt.ylabel('Number of People', color='black')
plt.show()
```

Here, we observe that there is mixed response from the passengers about the food and drinks that were served on the flight.

**3. Passenger Type vs Passenger Frequency**

- In the Bar Graph below, we have shown the number of passengers who chose to travel through the same airlines multiple times.

- We have plotted this graph to get an idea of how satisfied people were with the services of this particular airline and whether they choose to travel through this again.

```python
plt.figure(figsize = (7,5))
fig = sns.countplot(x = 'customer_type', data = df, palette = 'Paired', 
                    order = df['customer_type'].value_counts().index)
fig = plt.title('Bar graph for Type of Passenger vs Number of Passenger', fontsize = 15)
fig = plt.xlabel('Passenger Type', color='black')
fig = plt.ylabel('Number of Passengers', color='black')
plt.show()
```

- It can be inferred that most of the passengers were loyal and chose to travel through this particular airline multiple times signalling that their services were satisfying.

**4. Line graph of Ease of Online Booking vs Number of Passengers**
- In the Line Graph below, we have shown how easy/difficult it was for the passengers to book the tickets online for their flight.

- We have plotted this graph to get an idea of how satisfied people were with the online ticket booking services of this particular airline.

```python
grouped = df.groupby(['ease_of_online_booking']).count()
lp = sns.lineplot(x="ease_of_online_booking", y="gender", data=grouped);

lp.set(title='Line graph for Ease of Online Booking', xlabel='Rating scale(1 - 5)', ylabel='Number of passengers');
```

**5. Histogram of Age vs Number of Passengers**
- In the following plot, we have depicted the number of passengers of each age group who traveled through this particular airline.

```python
ax = df.hist(column='age', bins=7, grid=True, figsize=(7,5), color='#86bf91', zorder=2, rwidth=0.9)
ax = ax[0]
for x in ax:
    x.spines['right'].set_visible(False)
    x.spines['top'].set_visible(False)
    x.spines['left'].set_visible(False)
    
x.tick_params(axis="both", which="both", bottom="off", top="off", labelbottom="on", left="off", right="off", labelleft="on")

vals = x.get_yticks()
for tick in vals:
    x.axhline(y=tick, linestyle='dashed', alpha=0.4, color='#eeeeee', zorder=1)
    x.set_title(" Histogram of Age vs Number of Passengers")
    x.set_xlabel("Age", labelpad=20, weight='bold', size=12)
    x.set_ylabel("Number of Passenger", labelpad=20, weight='bold', size=12)
```

### Three-Variable Visualisation <a id='three_var'></a>

**1. Arrival Delay vs Departure Delay on the basis of Travel Purpose**
- In the Pair Plots below, we compared delay in arrival and departure on the basis of different ratings given by the passengers of this particular airline for Arrival Departure Time Convenience.
- The purpose behind making this plot is to check if arrival delay and departure delay are interdependent and how it influences the opinions of passengers with respect to Arrival Departure Time Convenience and whether it leads to major inconvenience or not. 

```python
sp = sns.pairplot(df,
                  vars=['arrival_delay', 'departure_delay'],
                  height=4,
                  hue='departure_arrival_time_convenience', palette='tab10');


# change the legend title
sp._legend.set_title('Travel Purpose'); 
sp.fig.suptitle('Arrival Delay vs Departure Delay on the basis of Travel Purpose', x=0.4,y=1.02);
```

- As we can see that the scatter plot is linear, we can infer that with the delay in departure of flights, the time for arrival delay also increases which further negatively impacts the ratings for Arrival Departure Time Convenience.

**2. Faceted Regression Plot for Age of Passenger vs Flight Distance on the basis of Passenger Travel Class**
- Through the following Faceted Regression Plot, we aim to understand the preferred travel class by the passengers of varying age groups for different flight distances.

```python
# Age and class and flight distance regression plot
g = sns.lmplot(x="flight_distance",
y="age",
hue="customer_class",
data = df,
palette = "magma");

# set axis labels
g.set_axis_labels("flight distance", "age");
g.fig.suptitle('Faceted Regression Plot for Age of Passenger vs Flight Distance on the basis of Passenger Travel Class',y=1.02);
```

- Here, we observe that for longer flight distances the adult passengers prefer to travel in business class.
- Youth and young adults prefer to mostly travel in economy or economy plus classes and take shorter flight routes.  

**3. Line plot for In-flight Entertainment vs Flight Distance on the basis of Gender**
- In the plot below, we want to illustrate how satisfied men and women are with the in-flight entertainment system on different duration of flights. 

```python
lp = sns.lineplot(x="inflight_entertainment",
y="flight_distance",
hue="gender",
data=df);

# position the legend outside the chart
lp.legend(bbox_to_anchor=(1.05, 1),
loc=2,
borderaxespad=0.0);

# add a title and axes labels
lp.set(title='Line plot of Inflight Entertainment vs Flight Distance and Gender',
xlabel='Inflight Entertainment',
ylabel='Flight Distance');
```

- Here we notice that ratings given by men for the in-flight entertainment system is slightly higher than women for long duration flights whereas the ratings by both men and women are very similar for short duration flights.

**4. Box plot for In-flight Wi-Fi Service vs In-flight Service on the basis of Travel Purpose**
- In the plot below, we have compared the in-flight wi-fi services and overall in-flight services for different travel purposes.
- The purpose of plotting this graph is to determine if there is a connection between the rating for in-flight wi-fi services with the overall in-flight services considering different purposes for which the passengers travel.

```python
# Travel purpose and wifi and in-flight service Box plot
plt.figure(figsize = (10,7))
lp = sns.boxplot(df['inflight_wifi_service'], df['inflight_service'],
                hue = df['travel_purpose'])

# position the legend outside the chart
lp.legend(bbox_to_anchor=(1.05, 1),
            loc=2,
            borderaxespad=0.0);

plt.title('Box plot for In-flight Wi-Fi Service and In-flight Service and Travel Purpose', fontsize = 15)
plt.xlabel('In-Flight Wi-Fi Service', color='black')
plt.ylabel('In-Flight Service', color='black')
plt.show();
```

Finally, it can be concluded from the plot that when the ratings for in-flight wi-fi services are good, then the overall wi-fi services for business are also averaged to be above 4. Similar trends were noticed for personal travel as well. Moreover, we can also note quite a few outliers for each rating.

## Literature Review <a id='literature'></a>

<br>

The research conducted portrays that the quality of the services provided by the airlines and the satisfaction level of the passengers has a direct impact on the preferences of the passengers for a particular airline. The outcomes of the research are helpful in managing the airlines as they illustrate the areas of improvements of the facilities provided to the passengers in order to satisfy more customers which indirectly results in increased number of customer flow. It was also found that the quality of the products, the price and the personal factors are three of the most predominant components in determining the satisfaction level of the passengers.
